```python
from mmedit.registry import MODELS
from mmedit.utils import register_all_modules
```

```python
model_cfg = model = dict(
    type='BaseEditModel',
    generator=dict(
        type='NAFNet',
        img_channel=3,
        mid_channels=64,
        enc_blk_nums=[2, 2, 4, 8],
        middle_blk_num=12,
        dec_blk_nums=[2, 2, 2, 2],
    ),
    pixel_loss=dict(type='PSNRLoss'),
    train_cfg=dict(),
    test_cfg=dict(),
    data_preprocessor=dict(
        type='EditDataPreprocessor',
        mean=[0., 0., 0.],
        std=[255., 255., 255.],
    )
)
```

```python
register_all_modules()
model = MODELS.build(model_cfg)
```

```python
from mmengine.runner import load_checkpoint
load_checkpoint(model, 'checkpoint/NAFNet-SIDD-midc64.pth')
```

```python
import torch
```

```python
img = torch.tensor(range(12)).view(1,3,2,2)/10*255
```

```python
import mmcv
img = mmcv.imread('../datasets/SIDD/val/noisy/val_166_NOISY.jpg')
```

```python
out = model(img)
```
